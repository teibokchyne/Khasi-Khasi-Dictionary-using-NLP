1. Download pdfs

```python
import internetarchive
import os
import time
from internetarchive import search_items

# Search for Khasi-language texts
search_results = internetarchive.search_items('language:"Kha" AND mediatype:texts')
first_10000 = list(search_results)[:10000]

# Create base directory
base_dir = 'downloads'
os.makedirs(base_dir, exist_ok=True)

# Track how many times each item has failed
retry_counts = {}

# Retry loop
max_retries = 5
pending = first_10000

while pending:
    next_pending = []

    for index, item in enumerate(pending):
        identifier = item['identifier']
        title = item.get('title', 'No Title')

        target_dir = os.path.join(base_dir, identifier)
        os.makedirs(target_dir, exist_ok=True)

        # Check if PDF already exists
        pdf_files = [f for f in os.listdir(target_dir) if f.lower().endswith('.pdf')]
        if pdf_files:
            print(f"✅ Skipping {identifier} (PDF already exists)")
            continue

        print(f"⬇️  Downloading ({index + 1}/{len(pending)}): {title} | ID: {identifier}")

        try:
            result = internetarchive.download(
                identifier,
                glob_pattern='*.pdf',
                destdir=target_dir
            )

            if not result:
                print(f"  → No PDFs found for '{identifier}'")

        except Exception as e:
            print(f"  → Failed to download {identifier}: {e}")
            retry_counts[identifier] = retry_counts.get(identifier, 0) + 1

            if retry_counts[identifier] <= max_retries:
                next_pending.append(item)
            else:
                print(f"  → Skipping {identifier} after {max_retries} failed attempts.")

        time.sleep(2)

    pending = next_pending
```

Clean the folders such that only text files pdf remain. That means the pdf which text can be highlighted. Not the pdf that consists of pictures scanned where text cannot be highlighted.

```python
import os
import shutil

base_dir = "downloads2"
destination = "downloads_ocr"

downloads_contents = os.listdir(base_dir)

for folder1 in downloads_contents:
    folder1_contents = os.listdir(os.path.join(base_dir,folder1))
    for folder2 in folder1_contents:
        folder2_contents = os.listdir(os.path.join(base_dir,folder1,folder2))
        for file in folder2_contents:
            if not file.endswith("text.pdf"):
                src = os.path.join(base_dir,folder1,folder2,file)
                des = destination
                shutil.move(src, des)
```

2. Download txt files

```python
import internetarchive
import os
import time
from internetarchive import search_items

# Search for Khasi-language texts
search_results = internetarchive.search_items('language:"Kha" AND mediatype:texts')
first_10000 = list(search_results)[:10000]

# Create base directory
base_dir = 'downloads_texts'
os.makedirs(base_dir, exist_ok=True)

# Track how many times each item has failed
retry_counts = {}

# Retry loop
max_retries = 5
pending = first_10000

while pending:
    next_pending = []

    for index, item in enumerate(pending):
        identifier = item['identifier']
        title = item.get('title', 'No Title')

        target_dir = os.path.join(base_dir, identifier)
        os.makedirs(target_dir, exist_ok=True)

        # Check if text file already exists
        text_files = [f for f in os.listdir(target_dir) if f.lower().endswith(('.txt', '.djvu.txt'))]
        if text_files:
            print(f"✅ Skipping {identifier} (Text file already exists)")
            continue

        print(f"⬇️  Downloading ({index + 1}/{len(pending)}): {title} | ID: {identifier}")

        try:
            # Download full text files (djvu.txt or .txt)
            result = internetarchive.download(
                identifier,
                glob_pattern='*.txt',
                destdir=target_dir
            )

            if not result:
                print(f"  → No text files found for '{identifier}'")

        except Exception as e:
            print(f"  → Failed to download {identifier}: {e}")
            retry_counts[identifier] = retry_counts.get(identifier, 0) + 1

            if retry_counts[identifier] <= max_retries:
                next_pending.append(item)
            else:
                print(f"  → Skipping {identifier} after {max_retries} failed attempts.")

        time.sleep(2)

    pending = next_pending
```
